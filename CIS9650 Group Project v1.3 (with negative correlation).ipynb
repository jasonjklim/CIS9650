# 3.	Correlation Analysis

#### Jason Lim

•	Find the pair with the strongest positive correlation (closest to 1)<br>

```python
returns=data.copy()
returns = returns.pct_change()
returns
```

```python
## Create Dataframe for correlation chart.

security_cor_df = returns.corr()

security_cor_df.head()
```

```python

# Create a function to eliminate correlation of itself
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

# Create a function to get highest correlation 
def get_top_correlations(df, n=5):
    au_corr = df.corr().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

```python
print(get_top_correlations(returns, 3))
```

## Strongest positive correlation is between BAC and JPM

```python
# Create a function to get lowest correlation 

def get_lowest_correlations(df, n=5):
    au_corr = df.corr().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values()
    return au_corr[0:n]
```

```python
print(get_lowest_correlations(returns, 3))
```

## Strongest negative correlation is between EDV and JPM

```python
# Heatmap
fig = plt.figure(figsize=(25,15))
ax=sns.heatmap(returns.corr(),annot=True, linewidths=0.5,cmap='coolwarm')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('Correlation Heatmap of 30 Securities', fontdict={'fontsize': 30, 'fontweight': 'heavy'})
```

```python
import numpy as np
```

```python
BAC=np.array(returns["BAC"])
JPM=np.array(returns["JPM"])
EDV=np.array(returns["EDV"])
```

# Regression Analysis: Strongest Positive Correlation (BAC & JPM)

```python
BAC = BAC[1:]
JPM = JPM[1:]
```

```python
#Converting array into vector
BAC = BAC.reshape(-1,1)
```

```python
#Converting array into vector
JPM = JPM.reshape(-1,1)
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
#To fit a linear model between BAC & JPM
model = reg.fit (BAC, JPM)
```

```python
#To find Y intercept
model.intercept_
```

```python
#To find slope
model.coef_
```

```python
# Y = 0.836x + 0.000
```

```python
#R^2
model.score (BAC, JPM)
```

```python
#Mean squared error
from sklearn.metrics import mean_squared_error
```

```python
# Prediction should be our X(independent variable), which is BAC.
y_pred = model.predict(BAC)
```

```python
mean_squared_error(JPM, y_pred)
```

```python
#To plot a scatter plot
plt.scatter (BAC, JPM, color = 'red')
#To plot a line graph
plt.plot (BAC, y_pred, color = 'black', linewidth = 3)

plt.xlabel('BAC Daily % Change')
plt.ylabel('JPM Daily % Change')
plt.title('BAC vs JPM prediction', fontdict={'fontsize': 20, 'fontweight': 'heavy'})
```

```python
# Final Result for 

r_sq = model.score(BAC, JPM)
print('coefficient of determination:', r_sq)
print('intercept: ',model.intercept_)

print('slope: ', model.coef_)
print("MSE: ", mean_squared_error(JPM, y_pred))
```
